```python
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np

# Function to create a random network
def random_network(N, P):
    G = nx.Graph()  # Create an undirected graph
    G.add_nodes_from(range(N))  # Add N nodes to the graph

    # Loop through each pair of nodes (i, j) to check if an edge should be added
    for i in range(N):
        for j in range(i + 1, N):  # This ensures we don't double-count edges (i, j) and (j, i)
            if random.random() < P:  # Generate a random number between 0 and 1
                G.add_edge(i, j)  # If the number is less than P, add an edge

    return G


```

```python
# Generate a random network with N=100 and P=0.01
G1 = random_network(100, 0.01)

# Visualize the adjacency matrix
adjacency_matrix = nx.to_numpy_array(G1)
plt.matshow(adjacency_matrix, cmap='Blues')  # Display matrix using 'Blues' colormap
plt.title("Adjacency Matrix of the Network")
plt.show()

# Visualize the network using a circular layout
pos = nx.circular_layout(G1)  # Circular layout for the nodes
nx.draw(G1, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=12)
plt.title("Random Network Visualization (N=100, P=0.01)")
plt.show()
```

```python
# Calculate the number of edges in the graph
num_edges = len(G1.edges())
print(f"Number of edges: {num_edges}")

# Calculate the average degree of the network
degrees = [G1.degree(node) for node in G1.nodes()]
average_degree = np.mean(degrees)
print(f"Average degree: {average_degree}")

# Calculate the expected number of edges: E[l] = P * (N*(N-1))/2
expected_edges = 0.01 * (100 * 99) / 2
print(f"Expected number of edges: {expected_edges}")
```

```python
# Generate another random network with N=100 and P=0.05
G2 = random_network(100, 0.05)

# Calculate number of edges and average degree for this network
num_edges_2 = len(G2.edges())
print(f"Number of edges (P=0.05): {num_edges_2}")

# Calculate the average degree for the second network
degrees_2 = [G2.degree(node) for node in G2.nodes()]
average_degree_2 = np.mean(degrees_2)
print(f"Average degree (P=0.05): {average_degree_2}")

# Visualize the network using a circular layout
pos_2 = nx.circular_layout(G2)
nx.draw(G2, pos_2, with_labels=True, node_size=500, node_color='lightcoral', font_size=12)
plt.title("Random Network Visualization (N=100, P=0.05)")
plt.show()

```

```python
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np

# Function to create a random network
def random_network(N, P):
    G = nx.Graph()  # Create an undirected graph
    G.add_nodes_from(range(N))  # Add N nodes to the graph

    # Loop through each pair of nodes (i, j) to check if an edge should be added
    for i in range(N):
        for j in range(i + 1, N):  # This ensures we don't double-count edges (i, j) and (j, i)
            if random.random() < P:  # Generate a random number between 0 and 1
                G.add_edge(i, j)  # If the number is less than P, add an edge

    return G

# Generate a random network with N=100 and P=0.01
G1 = random_network(100, 0.01)

# Visualize the adjacency matrix
adjacency_matrix = nx.to_numpy_array(G1)
plt.matshow(adjacency_matrix, cmap='Blues')  # Display matrix using 'Blues' colormap
plt.title("Adjacency Matrix of the Network")
plt.show()

# Visualize the network using a circular layout
pos = nx.circular_layout(G1)  # Circular layout for the nodes
nx.draw(G1, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=12)
plt.title("Random Network Visualization (N=100, P=0.01)")
plt.show()

# Calculate the number of edges in the graph
num_edges = len(G1.edges())
print(f"Number of edges: {num_edges}")

# Calculate the average degree of the network
degrees = [G1.degree(node) for node in G1.nodes()]
average_degree = np.mean(degrees)
print(f"Average degree: {average_degree}")

# Calculate the expected number of edges: E[l] = P * (N*(N-1))/2
expected_edges = 0.01 * (100 * 99) / 2
print(f"Expected number of edges: {expected_edges}")

# Now, let's generate another random network with N=100 and P=0.05
G2 = random_network(100, 0.05)

# Calculate number of edges and average degree for this network
num_edges_2 = len(G2.edges())
print(f"Number of edges (P=0.05): {num_edges_2}")

# Calculate the average degree for the second network
degrees_2 = [G2.degree(node) for node in G2.nodes()]
average_degree_2 = np.mean(degrees_2)
print(f"Average degree (P=0.05): {average_degree_2}")

# Randomly select two nodes U and V from the graph
nodes_list = list(G2.nodes())  # Convert the nodes to a list
U, V = random.sample(nodes_list, 2)  # Randomly sample 2 nodes from the list

# Calculate the shortest path between U and V
shortest_path = nx.shortest_path(G2, source=U, target=V)
print(f"Shortest path between nodes {U} and {V}: {shortest_path}")

# Visualize the network and highlight the shortest path calculated
edges = list(G2.edges())
path_edges = list(zip(shortest_path, shortest_path[1:]))

# Draw the network with the shortest path highlighted
pos_3 = nx.circular_layout(G2)
nx.draw(G2, pos_3, with_labels=True, node_size=500, node_color='lightgreen', font_size=12)
nx.draw_networkx_edges(G2, pos_3, edgelist=path_edges, edge_color='r', width=2)
plt.title(f"Random Network with Shortest Path between {U} and {V}")
plt.show()

# Generate a larger random network with N=1000 and P=0.005
G3 = random_network(1000, 0.005)

# Calculate number of edges and visualize the network
num_edges_3 = len(G3.edges())
print(f"Number of edges (P=0.005, N=1000): {num_edges_3}")

# Visualize the network (this will be large, so we'll sample it)
pos_4 = nx.spring_layout(G3, seed=42)  # Use a different layout to handle larger graphs
nx.draw(G3, pos_4, node_size=30, node_color='purple', font_size=8)
plt.title("Random Network Visualization (N=1000, P=0.005)")
plt.show()
```
